The following is an attempt at the classifcation of newswires. It is an example of a multiclass classification example.

````
#import keras datasets reuters
from keras.datasets import reuters
(train_data,train_labels),(test_data,test_labels) = reuters.load_data(num_words=10000)
````

````
#Encoding of data
import numpy as np
from tensorflow.keras.utils import to_categorical
#Vectorization of training data
#transform [1,2] => [0,1,1] etc but in our case we have 8982 sequences and have 10k possibilites
trainDataEncoder = np.zeros((len(train_data),10000))

def dataEncoder(data,numberOfWords):
  trainDataEncoder = np.zeros((len(data),numberOfWords))
  for rank,sequence in enumerate(data):
    #here we have the rank and the sequence , rank will tell us the first place
    for number in sequence:
      #we run through each number in sequence and add 1 to the place.
      trainDataEncoder[rank][number] = 1
  return trainDataEncoder

#return encoded data
trainX = dataEncoder(train_data,10000)
testX = dataEncoder(test_data,10000)
#return encoded labels
trainY = to_categorical(train_labels)
testY = to_categorical(test_labels)
````

````
#impliment and develop Model
from tensorflow import keras
from keras import models,layers
#build model
network = models.Sequential()
network.add(layers.Dense(256,input_shape=(10000,)))
network.add(layers.LeakyReLU())
network.add(layers.Dropout(0.5))
network.add(layers.Dense(256,input_shape=(10000,)))
network.add(layers.LeakyReLU())
network.add(layers.Dropout(0.5))
network.add(layers.Dense(128,input_shape=(10000,)))
network.add(layers.LeakyReLU())
network.add(layers.Dropout(0.5))
network.add(layers.Dense(64))
network.add(layers.LeakyReLU())
network.add(layers.BatchNormalization())
network.add(layers.Dense(46,activation='softmax'))
#compile model
network.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
#fit model
x_train = trainX[1000:]
x_val = trainX[:1000]
y_train = trainY[1000:]
y_val = trainY[:1000]
history = network.fit(x_train,y_train,epochs=100,batch_size=512,validation_data=(x_val,y_val))
````

````
from matplotlib import pyplot
#get accuracy
acc = np.asarray(history.history['accuracy'])
#get loss for testing
val_acc = np.asarray(history.history['val_accuracy'])
# x axis will display epochs running from 1 to the number of losses we have  - 1
epochs = np.asarray(range(len(acc)))
pyplot.plot(epochs,acc,'r',label='Training accuracy')
pyplot.plot(epochs,val_acc,'b',label='validation accuracy')
pyplot.title('Training and accuracy')
pyplot.xlabel('Epochs')
pyplot.ylabel('Wins')
pyplot.legend()
print(max(val_acc))
pyplot.show()
````
